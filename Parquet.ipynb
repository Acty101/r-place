```python
# convert int rgb to individual r g b values
# note r is 16^4 and 16^5, g is 16^2 and 16^3, b is 16^1 and 16^0
# input is a number, output is an array

def rgb_interpret(rgb):
    r1 = divmod(rgb, 16**5)
    r2 = divmod(r1[1], 16**4)
    g1 = divmod(r2[1], 16**3)
    g2 = divmod(g1[1],16**2)
    b1 = divmod(g2[1],16**1)
    b2 = divmod(b1[1],1)
     
    r = r1[0]*16 + r2[0]
    b = b1[0]*16 + b2[0]
    g = g1[0]* 16 + g2[0]
        
    return r,g,b
```
